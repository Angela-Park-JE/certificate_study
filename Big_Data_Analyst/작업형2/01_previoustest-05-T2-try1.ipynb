````python
# 3. Preprocessing(1) : dummy 화 성공
# help(pd.get_dummies)
# String to append DataFrame column names.

df_train_dum = pd.get_dummies(data = df_train, columns = ['transmission', 'fuelType'],
                              prefix = ['mission', 'fuel'], drop_first=False, dtype = np.int8)
df_train_dum.sample(5)
````

## 3. 전처리
### (2) train 데이터 test 데이터 더미화 후, X, y로 나누기

차종을 드랍하는 방식으로 간단하게 해보기.

````python
# 3. Preprocessing(2) : X, y 
# df_train_dum = pd.get_dummies(data = df_train, columns = ['transmission', 'fuelType'],
#                               prefix = ['mission', 'fuel'], drop_first=False, dtype = np.int8)

df_test_dum = pd.get_dummies(data = df_test, columns = ['transmission', 'fuelType'],
                              prefix = ['mission', 'fuel'], drop_first=False, dtype = np.int8)

X_train, y_train = df_train_dum.drop(columns = ['model', 'price']),  df_train_dum['price']
X_test = df_test_dum.drop(columns = ['model'])
````

## 4. 모델 생성
### (1) SVR

````python
# 1. SVR - kernel :rbf :5초걸렸나.
import sklearn.svm
# help(sklearn.svm.SVR)
svr_rbf = sklearn.svm.SVR(kernel = 'rbf', epsilon = 0.1)
# svr_rbf.fit(X = X_train, y = y_train)
# y_test = svr_rbf.predict(X_test)
````

````python
# # 1. SVR - kernel :linear : 5분이 넘어가서 중단하기로 함.
# # import sklearn.svm
# svr_lin = sklearn.svm.SVR(kernel = 'linear', epsilon = 0.1)
# svr_lin.fit(X = X_train, y = y_train)
# # y_test = svr_lin.predict(X_test)
````

### (2) Decision tree

````python
# 2. 디시젼트리 : friedman mean square errir
import sklearn.tree
# help(sklearn.tree.DecisionTreeRegressor)
tree = sklearn.tree.DecisionTreeRegressor(criterion = 'friedman_mse')
# tree.fit(X = X_train, y = y_train)
# y_test = tree.predict(X_test)
````

### (3) Gaussian

````python
# 3. 가우시안 : 3초걸렸나
import sklearn.gaussian_process
# help(sklearn.gaussian_process.GaussianProcessRegressor)
gaus = sklearn.gaussian_process.GaussianProcessRegressor()
# gaus.fit(X = X_train, y = y_train)
# y_test = gaus.predict(X_test)
````

### (4) 앙상블 - Ada boost

````python
# 4. adaboost : 3초도 안걸렸다는...
import sklearn.ensemble
# help(sklearn.ensemble.AdaBoostRegressor)
ada = sklearn.ensemble.AdaBoostRegressor()
# ada.fit(X = X_train, y = y_train)
# y_test = ada.predict(X_test)
````

### (5) 최종 - 위 모델들 Voting

```
 |  >>> import numpy as np
 |  >>> from sklearn.linear_model import LinearRegression
 |  >>> from sklearn.ensemble import RandomForestRegressor
 |  >>> from sklearn.ensemble import VotingRegressor
 |  >>> from sklearn.neighbors import KNeighborsRegressor
 |  >>> r1 = LinearRegression()
 |  >>> r2 = RandomForestRegressor(n_estimators=10, random_state=1)

 |  >>> y = np.array([2, 6, 12, 20, 30, 42])

```

````python
# 5. adaboost : 3초도 안걸렸다는...
import sklearn.ensemble
# help(sklearn.ensemble.VotingRegressor)
esb_voting = sklearn.ensemble.VotingRegressor([('svr', svr_rbf), ('dt', tree), ('ga', gaus), ('ada', ada)])
````

## 5. 예측

````python
# 5~6초 걸린 것 같음
esb_voting.fit(X = X_train, y = y_train)
pred = esb_voting.predict(X = X_test)
````

## 6. csv

````python
final = pd.DataFrame(pred)
final.reset_index(inplace = True)
final.to_csv('submission.csv', header = ['id', 'price'], index = False)
pd.read_csv('submission.csv')
````
